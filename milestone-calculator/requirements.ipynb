```python
def safe_calculator(num1, operator, num2):
    """å®‰å…¨çš„è¨ˆç®—æ©Ÿï¼ˆè™•ç†é™¤é›¶éŒ¯èª¤ï¼‰"""
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        if num2 == 0:
            return "éŒ¯èª¤ï¼šä¸èƒ½é™¤ä»¥é›¶ï¼"
        return num1 / num2
    else:
        return "éŒ¯èª¤ï¼šç„¡æ•ˆçš„é‹ç®—ç¬¦è™Ÿï¼"

# æ¸¬è©¦é™¤é›¶è™•ç†
print("é™¤é›¶éŒ¯èª¤è™•ç†æ¸¬è©¦ï¼š")
print(safe_calculator(10, "/", 0))   # éŒ¯èª¤ï¼šä¸èƒ½é™¤ä»¥é›¶ï¼
print(safe_calculator(10, "/", 2))   # 5.0
print(safe_calculator(0, "/", 5))    # 0.0ï¼ˆé›¶é™¤ä»¥å…¶ä»–æ•¸å­—æ˜¯å¯ä»¥çš„ï¼‰

# æ¸¬è©¦ç„¡æ•ˆé‹ç®—ç¬¦è™Ÿ
print("\nç„¡æ•ˆé‹ç®—ç¬¦è™Ÿæ¸¬è©¦ï¼š")
print(safe_calculator(10, "$", 5))   # éŒ¯èª¤ï¼šç„¡æ•ˆçš„é‹ç®—ç¬¦è™Ÿï¼
print(safe_calculator(10, "^", 5))   # éŒ¯èª¤ï¼šç„¡æ•ˆçš„é‹ç®—ç¬¦è™Ÿï¼
```

### éœ€æ±‚ 2.2ï¼šé‹ç®—ç¬¦è™Ÿæ“´å±•

**æè¿°**ï¼šæ”¯æ´æ›´å¤šé‹ç®—é¡å‹ã€‚

**è¦æ ¼**ï¼š
- æ¬¡æ–¹é‹ç®—ï¼ˆ**ï¼‰
- é¤˜æ•¸é‹ç®—ï¼ˆ%ï¼‰
- æ•´æ•¸é™¤æ³•ï¼ˆ//ï¼‰

**æŠ€è¡“å¯¦ä½œ**ï¼š

```python
def advanced_calculator(num1, operator, num2):
    """é€²éšè¨ˆç®—æ©Ÿï¼ˆæ”¯æ´æ›´å¤šé‹ç®—ï¼‰"""
    if operator == "+":
        return num1 + num2
    elif operator == "-":
        return num1 - num2
    elif operator == "*":
        return num1 * num2
    elif operator == "/":
        if num2 == 0:
            return "éŒ¯èª¤ï¼šä¸èƒ½é™¤ä»¥é›¶ï¼"
        return num1 / num2
    elif operator == "**":
        return num1 ** num2
    elif operator == "%":
        if num2 == 0:
            return "éŒ¯èª¤ï¼šä¸èƒ½å°é›¶æ±‚é¤˜æ•¸ï¼"
        return num1 % num2
    elif operator == "//":
        if num2 == 0:
            return "éŒ¯èª¤ï¼šä¸èƒ½é™¤ä»¥é›¶ï¼"
        return num1 // num2
    else:
        return "éŒ¯èª¤ï¼šç„¡æ•ˆçš„é‹ç®—ç¬¦è™Ÿï¼"

# æ¸¬è©¦é€²éšé‹ç®—
print("é€²éšé‹ç®—æ¸¬è©¦ï¼š")
print(f"2 ** 3 = {advanced_calculator(2, '**', 3)}")     # æ¬¡æ–¹ï¼š8
print(f"10 % 3 = {advanced_calculator(10, '%', 3)}")     # é¤˜æ•¸ï¼š1
print(f"10 // 3 = {advanced_calculator(10, '//', 3)}")   # æ•´æ•¸é™¤æ³•ï¼š3
print(f"7.5 % 2.5 = {advanced_calculator(7.5, '%', 2.5)}")  # å°æ•¸é¤˜æ•¸ï¼š0.0
```
