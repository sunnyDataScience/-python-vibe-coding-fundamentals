````python
# å–®ä¸€å›å‚³å€¼
def square(n):
    """è¨ˆç®—å¹³æ–¹"""
    return n ** 2

print(square(5))  # 25
print(square(10))  # 100
````

````python
# å¤šé‡å›å‚³å€¼ï¼ˆå¯¦éš›ä¸Šæ˜¯å›å‚³ tupleï¼‰
def divide_with_remainder(a, b):
    """è¨ˆç®—å•†å’Œé¤˜æ•¸"""
    quotient = a // b
    remainder = a % b
    return quotient, remainder  # å›å‚³å…©å€‹å€¼

# æ–¹æ³• 1ï¼šæ¥æ”¶ç‚º tuple
result = divide_with_remainder(17, 5)
print(f"Result: {result}")  # (3, 2)
print(f"Type: {type(result)}")  # <class 'tuple'>

# æ–¹æ³• 2ï¼šè§£åŒ…ï¼ˆUnpackingï¼‰
q, r = divide_with_remainder(17, 5)
print(f"Quotient: {q}, Remainder: {r}")  # Quotient: 3, Remainder: 2
````

````python
# æ¢ä»¶å¼å›å‚³
def get_grade(score):
    """æ ¹æ“šåˆ†æ•¸å›å‚³ç­‰ç¬¬"""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

print(get_grade(95))  # A
print(get_grade(75))  # C
print(get_grade(55))  # F
````

### ç¯„ä¾‹ 5ï¼šæ–‡ä»¶å­—ä¸²èˆ‡ help()

````python
# å®Œæ•´çš„ Docstring ç¯„ä¾‹
def calculate_bmi(weight, height):
    """
    è¨ˆç®—èº«é«”è³ªé‡æŒ‡æ•¸ï¼ˆBMIï¼‰
    
    åƒæ•¸:
        weight (float): é«”é‡ï¼ˆå…¬æ–¤ï¼‰
        height (float): èº«é«˜ï¼ˆå…¬å°ºï¼‰
    
    å›å‚³:
        float: BMI å€¼
    
    ç¯„ä¾‹:
        >>> calculate_bmi(70, 1.75)
        22.86
    """
    bmi = weight / (height ** 2)
    return round(bmi, 2)

# ä½¿ç”¨å‡½å¼
print(calculate_bmi(70, 1.75))
````

### ç¯„ä¾‹ 6ï¼šç´”å‡½å¼è¨­è¨ˆ

````python
# ç´”å‡½å¼ï¼ˆPure Functionï¼‰ï¼šè¼¸å‡ºåªä¾è³´è¼¸å…¥ï¼Œç„¡å‰¯ä½œç”¨
def add_pure(a, b):
    """ç´”å‡½å¼ï¼šè¨ˆç®—å…©æ•¸ä¹‹å’Œ"""
    return a + b

# ç›¸åŒè¼¸å…¥æ°¸é å¾—åˆ°ç›¸åŒè¼¸å‡º
print(add_pure(3, 5))  # 8
print(add_pure(3, 5))  # 8
print(add_pure(3, 5))  # 8
````

````python
# æœ‰å‰¯ä½œç”¨çš„å‡½å¼ï¼ˆä¿®æ”¹å¤–éƒ¨ç‹€æ…‹ï¼‰
counter = 0  # å…¨åŸŸè®Šæ•¸

def add_with_side_effect(a, b):
    """æœ‰å‰¯ä½œç”¨çš„å‡½å¼ï¼šä¿®æ”¹å…¨åŸŸè®Šæ•¸"""
    global counter
    counter += 1  # å‰¯ä½œç”¨ï¼šä¿®æ”¹å¤–éƒ¨è®Šæ•¸
    return a + b

print(add_with_side_effect(3, 5))  # 8
print(f"Counter: {counter}")  # 1

print(add_with_side_effect(3, 5))  # 8
print(f"Counter: {counter}")  # 2ï¼ˆå‰¯ä½œç”¨ç´¯ç©ï¼‰
````

````python
# âŒ éŒ¯èª¤ç¤ºç¯„ï¼šä½¿ç”¨å¯è®Šç‰©ä»¶ä½œç‚ºé è¨­å€¼
def add_item_wrong(item, items=[]):
    """éŒ¯èª¤ï¼šé è¨­åƒæ•¸æ˜¯å¯è®Šç‰©ä»¶"""
    items.append(item)
    return items

# å•é¡Œï¼šå¤šæ¬¡å‘¼å«å…±äº«åŒä¸€å€‹åˆ—è¡¨
list1 = add_item_wrong("apple")
print(f"List 1: {list1}")  # ['apple']

list2 = add_item_wrong("banana")
print(f"List 2: {list2}")  # ['apple', 'banana'] â† é æœŸæ‡‰è©²æ˜¯ ['banana']

list3 = add_item_wrong("cherry")
print(f"List 3: {list3}")  # ['apple', 'banana', 'cherry']
````

````python
# ä½¿ç”¨ id() é©—è­‰æ˜¯åŒä¸€å€‹ç‰©ä»¶
def add_item_debug(item, items=[]):
    print(f"List ID: {id(items)}")  # è¨˜æ†¶é«”ä½å€
    items.append(item)
    return items

add_item_debug("apple")
add_item_debug("banana")  # ç›¸åŒçš„ IDï¼
add_item_debug("cherry")
````

````python
# æ‡‰ç”¨ 1ï¼šæº«åº¦è½‰æ›å·¥å…·
def celsius_to_fahrenheit(celsius):
    """æ”æ°è½‰è¯æ°"""
    return celsius * 9/5 + 32

def fahrenheit_to_celsius(fahrenheit):
    """è¯æ°è½‰æ”æ°"""
    return (fahrenheit - 32) * 5/9

def celsius_to_kelvin(celsius):
    """æ”æ°è½‰å…‹æ°"""
    return celsius + 273.15

# æ¸¬è©¦
temp_c = 25
print(f"{temp_c}Â°C = {celsius_to_fahrenheit(temp_c):.1f}Â°F")
print(f"{temp_c}Â°C = {celsius_to_kelvin(temp_c):.2f}K")

temp_f = 77
print(f"{temp_f}Â°F = {fahrenheit_to_celsius(temp_f):.1f}Â°C")
````

````python
# æ‡‰ç”¨ 2ï¼šæˆç¸¾è™•ç†ç³»çµ±
def calculate_average(scores):
    """è¨ˆç®—å¹³å‡åˆ†æ•¸"""
    if not scores:
        return 0
    return sum(scores) / len(scores)

def get_letter_grade(score):
    """åˆ†æ•¸è½‰ç­‰ç¬¬"""
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

def is_passing(score, passing_score=60):
    """åˆ¤æ–·æ˜¯å¦åŠæ ¼"""
    return score >= passing_score

def analyze_scores(scores):
    """
    åˆ†ææˆç¸¾çµ±è¨ˆ
    
    å›å‚³:
        dict: åŒ…å«å¹³å‡ã€æœ€é«˜ã€æœ€ä½ã€åŠæ ¼äººæ•¸
    """
    avg = calculate_average(scores)
    passing_count = sum(1 for s in scores if is_passing(s))
    
    return {
        "average": round(avg, 2),
        "highest": max(scores) if scores else 0,
        "lowest": min(scores) if scores else 0,
        "passing_count": passing_count,
        "total_count": len(scores)
    }

# æ¸¬è©¦
student_scores = [85, 92, 78, 65, 88, 73, 95, 58]
print(f"å¹³å‡åˆ†æ•¸: {calculate_average(student_scores):.2f}")
print(f"ç­‰ç¬¬: {get_letter_grade(85)}")
print(f"æ˜¯å¦åŠæ ¼: {is_passing(75)}")
print(f"\næˆç¸¾åˆ†æ:")
print(analyze_scores(student_scores))
````

````python
# æ‡‰ç”¨ 3ï¼šå­—ä¸²è™•ç†å·¥å…·
def capitalize_words(text):
    """å°‡æ¯å€‹å–®å­—é¦–å­—æ¯å¤§å¯«"""
    return text.title()

def count_words(text):
    """è¨ˆç®—å–®å­—æ•¸é‡"""
    return len(text.split())

def reverse_string(text):
    """åè½‰å­—ä¸²"""
    return text[::-1]

def is_palindrome(text):
    """æª¢æŸ¥æ˜¯å¦ç‚ºè¿´æ–‡"""
    cleaned = text.lower().replace(" ", "")
    return cleaned == cleaned[::-1]

# æ¸¬è©¦
sentence = "hello world"
print(capitalize_words(sentence))
print(f"Word count: {count_words(sentence)}")
print(f"Reversed: {reverse_string(sentence)}")
print(f"Is 'racecar' a palindrome? {is_palindrome('racecar')}")
print(f"Is 'hello' a palindrome? {is_palindrome('hello')}")
````
