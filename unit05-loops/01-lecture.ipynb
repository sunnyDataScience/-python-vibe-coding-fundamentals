## ç¯„ä¾‹ 1ï¼šfor è¿´åœˆåŸºæœ¬ç”¨æ³•

````python
# å°å‡º 0 åˆ° 4
for i in range(5):
    print(f"ç¬¬ {i} æ¬¡è¿­ä»£")

print("è¿´åœˆçµæŸ")
````

## ç¯„ä¾‹ 2ï¼šrange() ä¸‰ç¨®ç”¨æ³•å°æ¯”

````python
# ç”¨æ³• 1: range(stop)
print("range(5):")
for i in range(5):
    print(i, end=" ")
print()

# ç”¨æ³• 2: range(start, stop)
print("\nrange(1, 6):")
for i in range(1, 6):
    print(i, end=" ")
print()

# ç”¨æ³• 3: range(start, stop, step)
print("\nrange(0, 10, 2):")
for i in range(0, 10, 2):
    print(i, end=" ")
print()

# è² æ•¸æ­¥é€²ï¼ˆå€’æ•¸ï¼‰
print("\nrange(5, 0, -1):")
for i in range(5, 0, -1):
    print(i, end=" ")
````

## ç¯„ä¾‹ 3ï¼šéæ­·å­—ä¸²

````python
# for è¿´åœˆå¯ä»¥éæ­·ä»»ä½•ã€Œå¯è¿­ä»£ç‰©ä»¶ã€
text = "Python"

print("é€å­—å…ƒåˆ—å°ï¼š")
for char in text:
    print(char)

print("\nå¸¶ç´¢å¼•çš„éæ­·ï¼š")
for i in range(len(text)):
    print(f"ç´¢å¼• {i}: {text[i]}")
````

## ç¯„ä¾‹ 4ï¼šwhile è¿´åœˆåŸºæœ¬ç”¨æ³•

````python
# å°å‡º 0 åˆ° 4ï¼ˆç”¨ while å¯¦ä½œï¼‰
count = 0
while count < 5:
    print(f"count = {count}")
    count += 1  # é‡è¦ï¼å¿…é ˆæ›´æ–°æ¢ä»¶

print(f"è¿´åœˆçµæŸï¼Œcount = {count}")
````

## ç¯„ä¾‹ 5ï¼šbreak - æå‰ä¸­æ–·è¿´åœˆ

````python
# æ‰¾åˆ°ç¬¬ä¸€å€‹å¶æ•¸å°±åœæ­¢
numbers = [1, 3, 5, 8, 9, 11]

for num in numbers:
    if num % 2 == 0:
        print(f"æ‰¾åˆ°ç¬¬ä¸€å€‹å¶æ•¸ï¼š{num}")
        break  # ç«‹å³è·³å‡ºè¿´åœˆ
    print(f"{num} æ˜¯å¥‡æ•¸")

print("è¿´åœˆçµæŸ")
````

## ç¯„ä¾‹ 6ï¼šcontinue - è·³éæœ¬æ¬¡è¿­ä»£

````python
# åªå°å‡ºå¥‡æ•¸
for i in range(10):
    if i % 2 == 0:
        continue  # è·³éå¶æ•¸ï¼Œç›´æ¥é€²å…¥ä¸‹æ¬¡è¿­ä»£
    print(i, end=" ")

print("\nè¿´åœˆçµæŸ")
````

## ç¯„ä¾‹ 7ï¼šelse å­å¥ - è¿´åœˆæ­£å¸¸çµæŸæ™‚åŸ·è¡Œ

````python
# åœ¨åˆ—è¡¨ä¸­æœå°‹ç›®æ¨™å€¼
numbers = [1, 3, 5, 7, 9]
target = 4

for num in numbers:
    if num == target:
        print(f"æ‰¾åˆ° {target}ï¼")
        break
else:
    # åªæœ‰æ²’æœ‰åŸ·è¡Œ break æ™‚æ‰æœƒåŸ·è¡Œ
    print(f"æ²’æœ‰æ‰¾åˆ° {target}")

# è©¦è©¦çœ‹æ”¹æˆ target = 5ï¼Œè§€å¯Ÿå·®ç•°
````

## ç¯„ä¾‹ 8ï¼šå·¢ç‹€è¿´åœˆ - ä¹ä¹ä¹˜æ³•è¡¨

````python
# å¤–å±¤è¿´åœˆï¼šæ§åˆ¶åˆ—ï¼ˆ1-9ï¼‰
for i in range(1, 10):
    # å…§å±¤è¿´åœˆï¼šæ§åˆ¶æ¬„ï¼ˆ1-9ï¼‰
    for j in range(1, 10):
        print(f"{i}x{j}={i*j:2d}", end="  ")
    print()  # æ¯å°å®Œä¸€åˆ—å°±æ›è¡Œ
````

## ç¯„ä¾‹ 9ï¼šç´¯åŠ å™¨æ¨¡å¼ (Accumulator Pattern)

````python
# è¨ˆç®— 1 åˆ° 100 çš„ç¸½å’Œ
total = 0  # ç´¯åŠ å™¨åˆå§‹å€¼

for i in range(1, 101):
    total += i  # ç´¯åŠ 

print(f"1 åˆ° 100 çš„ç¸½å’Œ = {total}")

# é©—è­‰ï¼šé«˜æ–¯å…¬å¼ n*(n+1)/2
print(f"é«˜æ–¯å…¬å¼é©—è­‰ï¼š{100*101//2}")
````

## ç¯„ä¾‹ 10ï¼šæ——æ¨™è®Šæ•¸ (Flag Variable)

````python
# æª¢æŸ¥æ•¸å­—æ˜¯å¦ç‚ºè³ªæ•¸
num = 17
is_prime = True  # æ——æ¨™ï¼šå‡è¨­æ˜¯è³ªæ•¸

for i in range(2, num):
    if num % i == 0:
        is_prime = False  # æ‰¾åˆ°å› æ•¸ï¼Œä¸æ˜¯è³ªæ•¸
        break

if is_prime:
    print(f"{num} æ˜¯è³ªæ•¸")
else:
    print(f"{num} ä¸æ˜¯è³ªæ•¸")
````

## ç¯„ä¾‹ 11ï¼šç„¡çª®è¿´åœˆèˆ‡é˜²è­·æ©Ÿåˆ¶

````python
# âš ï¸ å±éšªï¼šç„¡çª®è¿´åœˆç¯„ä¾‹ï¼ˆä¸è¦åŸ·è¡Œï¼ï¼‰
# x = 0
# while x < 10:
#     print(x)  # å¿˜è¨˜æ›´æ–° xï¼Œæ°¸é ä¸æœƒåœæ­¢ï¼

# âœ… å®‰å…¨ï¼šåŠ å…¥é˜²è­·æ©Ÿåˆ¶
count = 0
max_iterations = 1000  # æœ€å¤§è¿­ä»£æ¬¡æ•¸

while True:
    count += 1
    print(f"è¿­ä»£ {count}")
    
    if count >= 5:  # å¯¦éš›åœæ­¢æ¢ä»¶
        print("æ­£å¸¸çµæŸ")
        break
    
    if count > max_iterations:  # é˜²è­·æ©Ÿåˆ¶
        print("é”åˆ°æœ€å¤§è¿­ä»£æ¬¡æ•¸ï¼Œå¼·åˆ¶ä¸­æ–·")
        break
````

## ç¯„ä¾‹ 12ï¼šfor vs while å°æ¯” - åŒä¸€å•é¡Œå…©ç¨®è§£æ³•

````python
# å•é¡Œï¼šè¨ˆç®— 1 åˆ° 10 çš„éšä¹˜ (1*2*3*...*10)

# è§£æ³• 1ï¼šä½¿ç”¨ for
factorial_for = 1
for i in range(1, 11):
    factorial_for *= i
print(f"for è¿´åœˆçµæœï¼š{factorial_for}")

# è§£æ³• 2ï¼šä½¿ç”¨ while
factorial_while = 1
i = 1
while i <= 10:
    factorial_while *= i
    i += 1
print(f"while è¿´åœˆçµæœï¼š{factorial_while}")

# çµè«–ï¼šfor æ›´ç°¡æ½”ï¼Œå› ç‚ºè¿­ä»£æ¬¡æ•¸å·²çŸ¥
````
