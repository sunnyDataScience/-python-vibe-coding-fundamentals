### ğŸ’¡ ç¯„ä¾‹ 3ï¼šé‹ç®—å­å„ªå…ˆé †åºï¼ˆOperator Precedenceï¼‰

````python
# ç¯„ä¾‹ 1ï¼šä¹˜æ³•å„ªå…ˆæ–¼åŠ æ³•
result1 = 2 + 3 * 4
print(f"2 + 3 * 4 = {result1}")  # 14 (å…ˆç®— 3*4=12, å†ç®— 2+12)

# ç”¨æ‹¬è™Ÿæ”¹è®Šé †åº
result2 = (2 + 3) * 4
print(f"(2 + 3) * 4 = {result2}")  # 20 (å…ˆç®— 2+3=5, å†ç®— 5*4)

# ç¯„ä¾‹ 2ï¼šæŒ‡æ•¸å„ªå…ˆæ–¼ä¹˜æ³•
result3 = 2 * 3 ** 2
print(f"\n2 * 3 ** 2 = {result3}")  # 18 (å…ˆç®— 3**2=9, å†ç®— 2*9)

# ç”¨æ‹¬è™Ÿæ”¹è®Šé †åº
result4 = (2 * 3) ** 2
print(f"(2 * 3) ** 2 = {result4}")  # 36 (å…ˆç®— 2*3=6, å†ç®— 6**2)
````

### ğŸ’¡ ç¯„ä¾‹ 4ï¼šæ¯”è¼ƒé‹ç®—å­ï¼ˆComparison Operatorsï¼‰

æ¯”è¼ƒé‹ç®—å­ç”¨æ–¼æ¯”è¼ƒå…©å€‹å€¼ï¼Œå›å‚³ True æˆ– False

````python
# å…­ç¨®æ¯”è¼ƒé‹ç®—å­
a = 10
b = 5

print(f"a = {a}, b = {b}")
print(f"a > b  (å¤§æ–¼):     {a > b}")   # True
print(f"a < b  (å°æ–¼):     {a < b}")   # False
print(f"a >= b (å¤§æ–¼ç­‰æ–¼): {a >= b}")  # True
print(f"a <= b (å°æ–¼ç­‰æ–¼): {a <= b}")  # False
print(f"a == b (ç­‰æ–¼):     {a == b}")  # False
print(f"a != b (ä¸ç­‰æ–¼):   {a != b}")  # True
````

**æ³¨æ„äº‹é …**ï¼š

````python
# 1. int èˆ‡ float å¯ä»¥æ¯”è¼ƒ
print(f"5 == 5.0: {5 == 5.0}")  # True

# 2. å­—ä¸²æŒ‰å­—å…¸åºï¼ˆlexicographicï¼‰æ¯”è¼ƒ
print(f"\n'apple' < 'banana': {'apple' < 'banana'}")  # True
print(f"'10' < '9': {'10' < '9'}")  # Trueï¼ˆæ³¨æ„ï¼šå­—ä¸²æ¯”è¼ƒï¼ï¼‰

# 3. Python ç‰¹æœ‰ï¼šé€£é–æ¯”è¼ƒ
age = 25
print(f"\n18 <= {age} < 65: {18 <= age < 65}")  # True
# ç­‰åŒæ–¼ï¼š(18 <= age) and (age < 65)
````

### ğŸ’¡ ç¯„ä¾‹ 5ï¼šé‚è¼¯é‹ç®—å­ï¼ˆLogical Operatorsï¼‰

````python
# andï¼šå…©è€…çš†çœŸæ‰ç‚ºçœŸ
print("=== and é‹ç®—å­ ===")
print(f"True and True:   {True and True}")    # True
print(f"True and False:  {True and False}")   # False
print(f"False and False: {False and False}")  # False

# orï¼šä¸€è€…ç‚ºçœŸå³ç‚ºçœŸ
print("\n=== or é‹ç®—å­ ===")
print(f"True or True:   {True or True}")      # True
print(f"True or False:  {True or False}")     # True
print(f"False or False: {False or False}")    # False

# notï¼šå–å
print("\n=== not é‹ç®—å­ ===")
print(f"not True:  {not True}")   # False
print(f"not False: {not False}")  # True
````

**å¯¦éš›æ‡‰ç”¨**ï¼š

````python
# ç¯„ä¾‹ï¼šæª¢æŸ¥è³‡æ ¼
age = 20
has_license = True

# æ¢ä»¶ï¼šå¹´æ»¿ 18 æ­²ä¸”æœ‰é§•ç…§
can_drive = (age >= 18) and has_license
print(f"å¹´é½¡: {age}, æœ‰é§•ç…§: {has_license}")
print(f"å¯ä»¥é–‹è»Šå—ï¼Ÿ{can_drive}")

# æ¢ä»¶ï¼šæœªæˆå¹´æˆ–æ²’é§•ç…§
cannot_drive = (age < 18) or (not has_license)
print(f"ä¸èƒ½é–‹è»Šå—ï¼Ÿ{cannot_drive}")
````

### ğŸ’¡ ç¯„ä¾‹ 6ï¼šçŸ­è·¯æ±‚å€¼ï¼ˆShort-circuit Evaluationï¼‰

Python çš„ and/or é‹ç®—å­æœƒåœ¨ç¢ºå®šçµæœå¾Œç«‹å³åœæ­¢é‹ç®—

````python
# and çŸ­è·¯ï¼šç¬¬ä¸€å€‹ç‚º False æ™‚ï¼Œä¸æª¢æŸ¥ç¬¬äºŒå€‹
print("=== and çŸ­è·¯æ±‚å€¼ ===")
x = 5
result = (x < 0) and (10 / x > 2)  # x < 0 ç‚º Falseï¼Œä¸åŸ·è¡Œ 10/x
print(f"(x < 0) and (10 / x > 2) = {result}")

# or çŸ­è·¯ï¼šç¬¬ä¸€å€‹ç‚º True æ™‚ï¼Œä¸æª¢æŸ¥ç¬¬äºŒå€‹
print("\n=== or çŸ­è·¯æ±‚å€¼ ===")
result = (x > 0) or (10 / x > 2)  # x > 0 ç‚º Trueï¼Œä¸åŸ·è¡Œ 10/x
print(f"(x > 0) or (10 / x > 2) = {result}")
````

````python
# ç¯„ä¾‹ï¼šBMI è¨ˆç®—èˆ‡åˆ¤æ–·
weight = 70  # å…¬æ–¤
height = 1.75  # å…¬å°º

# è¨ˆç®— BMI
bmi = weight / (height ** 2)
print(f"BMI: {bmi:.2f}")

# åˆ¤æ–· BMI ç¯„åœ
is_underweight = bmi < 18.5
is_normal = 18.5 <= bmi < 24
is_overweight = 24 <= bmi < 27
is_obese = bmi >= 27

print(f"\né«”é‡éè¼•: {is_underweight}")
print(f"æ­£å¸¸ç¯„åœ: {is_normal}")
print(f"éé‡: {is_overweight}")
print(f"è‚¥èƒ–: {is_obese}")
````
